```python
import pandas as pd
%matplotlib inline
```

```python
df=pd.read_csv('https://github.com/prasertcbs/tutorial/raw/master/mpg.csv')
df.sample(5)
```

```python
df.plot.scatter(x='displ', y='cty')
```

```python
%config InlineBackend.figure_format = 'retina'
```

```python
df.plot.scatter(x='displ', y='cty')
```

```python
df.plot.scatter(x='displ', y='cty', grid=True, title='City miles per gallon vs. engine size (litres)')
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                title='City miles per gallon vs. engine size (litres)');
```

```python
df.sample(5)
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),
                title='City miles per gallon vs. engine size (litres)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),
                s=df.displ*10,
                title='City miles per gallon vs. engine size (litres)');
```

```python
df.sample(5)
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),
                s=df.cyl*10,
                title='City miles per gallon vs. engine size (litres)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),
                s=df.cyl*10,
                alpha=.5,
                title='City miles per gallon vs. engine size (litres)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),
                s=30,
                alpha=.5,
                title='City miles per gallon vs. engine size (litres)');
```

```python
df.plot.scatter(x='displ', y='cty', 
                grid=True, 
                c=df.drv.map({'4':'green', 'f':'blue', 'r':'red'}),
                s=df.cyl*10,
                alpha=.5,
                ylim=(0, 40),
                xlim=(0, 10),
                title='City miles per gallon vs. engine size (litres)');
```
