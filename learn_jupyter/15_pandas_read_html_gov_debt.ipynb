```python
dfs=pd.read_html('https://en.wikipedia.org/wiki/Government_debt', 
                 match='country', 
                 header=0)
len(dfs)
```

```python
df=dfs[0]
df
```

```python
df.info()
```

```python
df['% of GDP']=df['% of GDP'].str.replace(r'%', '').astype('float')
```

```python
df['% of world public debt']=df['% of world public debt'].str.replace(r'%', '').astype('float')
```

```python
df
```

```python
df.info()
```

```python
df.country=df.country.str.replace(r'[^\s+\w]', '')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
plt.figure(figsize=(7, 7))
sns.barplot(x='% of GDP', y='country', 
            data=df.sort_values(by='% of GDP', ascending=False),
            color='.75')
```

```python
dg=df.sort_values(by='% of GDP', ascending=False)
dg
```

```python
colors = ['deepskyblue' if s == 'Thailand' else '.75' for s in dg['country']]
colors
```

```python
plt.figure(figsize=(7, 7))
sns.barplot(x='% of GDP', y='country', 
            data=df.sort_values(by='% of GDP', ascending=False),
            palette=colors);
```

```python
colors = ['deepskyblue' if s == 'Thailand' else '.75' if s != 'World' else 'yellowgreen' for s in dg['country']]
colors
```

```python
plt.figure(figsize=(7, 7))
sns.barplot(x='% of GDP', y='country', 
            data=df.sort_values(by='% of GDP', ascending=False),
            palette=colors)
sns.despine()
plt.ylabel('');
```
