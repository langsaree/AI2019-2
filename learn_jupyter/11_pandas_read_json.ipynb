## Method 2: use OrderedDict

```python
from collections import OrderedDict
import json
import requests
```

```python
url = 'https://github.com/prasertcbs/test/raw/master/NineTemples.json'
r = requests.get(url)
r.text
```

```python
j = json.loads(r.text, object_pairs_hook=OrderedDict)
print(json.dumps(j, indent=2, sort_keys=False))
```

```python
df = pd.DataFrame(j)
df
```
