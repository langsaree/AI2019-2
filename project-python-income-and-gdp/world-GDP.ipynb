```python
import pandas as pd
df = pd.read_html('https://en.wikipedia.org/wiki/Government_debt')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df['% of GDP']=df['% of GDP'].str.replace(r'%', '').astype('float')
df['% of world public debt']=df['% of world public debt'].str.replace(r'%', '').astype('float')
```

```python
df.country=df.country.str.replace(r'[^\s+\w]', '')
```

```python
df
```

```python
plt.figure(figsize=(7, 7))
sns.barplot(x='% of GDP', y='country', 
            data=df.sort_values(by='% of GDP', ascending=False),
            color='.75')
```
